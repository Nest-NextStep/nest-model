```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)

# CHECK THE CORRELATION BETWEEN ALL FEATURES WITH LABEL (MAJOR)
traincorr_df = trainclean_df.copy()

# Encode the 'major' column
traincorr_df['Major_encoded'] = LabelEncoder().fit_transform(traincorr_df['major'])

# Select only numeric columns and calculate the correlation matrix
numeric_df = traincorr_df.select_dtypes(include='number')
correlation_matrix = numeric_df.corr()

# Extract the correlation with 'Major_encoded'
correlation_with_major = correlation_matrix['Major_encoded']

# Filtering out columns that have a correlation of zero with 'Major_encoded'
correlation_with_major = correlation_with_major[correlation_with_major != 0]

# Ordering the correlations by their absolute values from most to least
correlation_with_major_sorted = correlation_with_major.reindex(correlation_with_major.abs().sort_values(ascending=False).index)
print(correlation_with_major_sorted)

# Plot a heatmap for the correlation matrix
heatmap_data = correlation_matrix[['Major_encoded']]

# Plot a heatmap for the correlation with 'Major_encoded' only
plt.figure(figsize=(10, 24))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt=".2f", cbar_kws={'label': 'Correlation'})
plt.title('Correlation with Major_encoded')
plt.show()
```

### CLEANING MAJOR

```python
# DROP EMPTY STRING

trainmajor_df = trainmajor_df[trainmajor_df['major'] != '']
trainmajor_df.reset_index(drop=True, inplace=True)
print("Drop empty string successful!")
```

```python
# DECODE, CHANGE & AND / AND COMMA, REMOVE NOISE, UNUSED SPACE

import re
from unidecode import unidecode

trainmajor_df['major'] = trainmajor_df['major'].apply(unidecode)
trainmajor_df['major'] = trainmajor_df['major'].str.replace('&', ' and ')
trainmajor_df['major'] = trainmajor_df['major'].str.replace('/', ' or ')
trainmajor_df['major'] = trainmajor_df['major'].str.replace(',', ' or ')
trainmajor_df['major'] = trainmajor_df['major'].str.replace(r'[\W\d_]+', ' ', regex=True)
trainmajor_df['major'] = trainmajor_df['major'].str.lower().str.strip().str.replace(r'\s+', ' ', regex=True)
```

```python
# DROP ROW WITH UNFIXED MAJOR (IT CHECKS INDEPENDENT WORD)

import re

words_to_delete = ['or', 'dont', 'will', 'undecided', 'not', 'no', 'yet']
pattern = rf'\b({"|".join(words_to_delete)})\b'
trainmajor_df = trainmajor_df[~trainmajor_df['major'].str.contains(pattern, regex=True, case=False)]
trainmajor_df.reset_index(drop=True, inplace=True)
```

```python
# LEMMATIZER FOR SIMPLER THE WORD

import pandas as pd
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('wordnet')

def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(lemmatized_tokens)

trainmajor_df['major'] = trainmajor_df['major'].apply(lemmatize_text)
print("Successful for lemmatizing!")
```

```python
# DELETE WORD UNDER 2 LETTERS AND MANUAL DELETATION

import pandas as pd
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: ' '.join([word for word in x.split() if len(word) > 1]))

custom_stopwords = ['aa', 'an', 'aas', 'applied', 'associate', 'attending', 'ba', 'bachelor', 'bachelors', 'currently', 'college', 'diploma', 'degree', 'in', 'know', 'major', 'non', 'none', 'school', 'still', 'studies', 'study', 'the', 'studying', 'university', 'year', 'years', 'yes']
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in custom_stopwords]))
```

```python
# CHECK TYPO (IT CHECKS INDEPENDENT WORDS)

major_mapping = {
    r'\bengineer\b': 'engineering',
    r'\bit\b': 'information technology',
    r'\btech\b': 'technology',
    r'\bcom\b': 'computer',
    r'\baccountancy\b': 'accounting',
    r'\beconomy\b': 'economic',
    r'\beng\b': 'engineering',
    r'\bcomm\b': 'communication',
    r'\bhr\b': 'human resource'
}

def correct_major(major):
    for typo, correct in major_mapping.items():
        major = re.sub(typo, correct, major, flags=re.IGNORECASE)
    return major

trainmajor_df['major'] = trainmajor_df['major'].apply(correct_major)
```

```python
# CHECK TYPO (IT ALSO CHECKS WORD IN WORD)

def replace_word(sentence, target_substring, replacement):
    words = sentence.split()
    replaced_words = [replacement if target_substring in word else word for word in words]
    return ' '.join(replaced_words)

# Applying the transformation to the 'major' column
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: replace_word(x, 'acco', 'accounting'))
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: replace_word(x, 'adm', 'administration'))
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: replace_word(x, 'ant', 'anthropology'))
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: replace_word(x, 'archi', 'architecture'))
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: replace_word(x, 'bus', 'business'))
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: replace_word(x, 'vet', 'veterinary'))
# trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: replace_word(x, 'ant', 'anthropology'))
```

```python
# GROUPING MAJOR IN SIMILIAR (IT ALSO CHECK WORD IN WORD)

trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: 'architecture' if 'architecture' in x else x)
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: 'chemical engineering' if 'chemical eng' in x else x)
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: 'civil engineering' if 'civil' in x else x)
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: 'computer science' if 'computer science' in x else x)
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: 'electrical engineering' if 'electric' in x else x)
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: 'information systems' if 'information system' in x else x)
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: 'information technology' if 'information technology' in x else x)
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: 'interior design' if 'interior' in x else x)
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: 'mathematics' if 'math' in x else x)
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: 'mechanical engineering' if 'mechanical' in x else x)
trainmajor_df['major'] = trainmajor_df['major'].apply(lambda x: 'psychology' if 'psycho' in x else x)
```
