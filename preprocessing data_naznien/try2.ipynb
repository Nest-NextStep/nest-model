```python
import pandas as pd
df = pd.read_csv('data//preprocessed_data.csv')
df
```

```python
df= df.dropna(subset=['major'])
df= df.drop(columns=['R3'])
df= df.drop(columns=['R5'])
df= df.drop(columns=['I3'])
df= df.drop(columns=['I6'])
df= df.drop(columns=['A7'])
df= df.drop(columns=['A1'])
df= df.drop(columns=['S2'])
df= df.drop(columns=['S4'])
df= df.drop(columns=['E2'])
df= df.drop(columns=['E6'])
df= df.drop(columns=['C4'])
df= df.drop(columns=['C1'])
df= df.drop(columns=['VCL9'])
df= df.drop(columns=['VCL7'])
df= df.drop(columns=['VCL8'])
df= df.drop(columns=['VCL16'])
df= df.drop(columns=['hand'])
df= df.drop(columns=['age'])
df= df.drop(columns=['familysize'])
df= df.drop(columns=['urban'])
df= df.drop(columns=['race'])
df= df.drop(columns=['married'])
df= df.drop(columns=['orientation'])

```

```python
df['R'] = df[['R1', 'R2', 'R4', 'R6', 'R7', 'R8']].sum(axis=1)
df['I'] = df[['I1', 'I2', 'I4', 'I5', 'I7', 'I8']].sum(axis=1)
df['A'] = df[['A3', 'A2', 'A4', 'A5', 'A6', 'A8']].sum(axis=1)
df['S'] = df[['S1', 'S3', 'S5', 'S6', 'S7', 'S8']].sum(axis=1)
df['E'] = df[['E1', 'E3', 'E4', 'E5', 'E7', 'E8']].sum(axis=1)
df['C'] = df[['C3', 'C2', 'C5', 'C6', 'C7', 'C8']].sum(axis=1)
df
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

# Misalkan df adalah DataFrame yang berisi data
# Contoh inisialisasi data
# df = pd.DataFrame({...})

# Atribut "major"
major_column = 'major'

# Jumlah cluster yang diinginkan
n_clusters = 60

# Melakukan clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(df.drop(columns=[major_column]))

# Menambahkan informasi cluster ke dalam DataFrame
df['cluster'] = clusters

# Menyimpan hasil analisis
results = []

# Menyimpan cluster yang sudah digunakan
used_clusters = set()

# Mendapatkan nilai unik dari atribut "major"
unique_majors = df[major_column].unique()

for major in unique_majors:
    # Mendapatkan data untuk nilai "major" tertentu
    df_major = df[df[major_column] == major]
    
    # Menghitung jumlah anggota di setiap cluster untuk "major" ini
    cluster_counts = df_major['cluster'].value_counts()
    
    # Menemukan cluster yang belum digunakan dengan anggota terbanyak
    for cluster, count in cluster_counts.items():
        if cluster not in used_clusters:
            most_common_cluster = cluster
            most_common_cluster_count = count
            used_clusters.add(cluster)
            break
    
    results.append({
        'major': major,
        'most_common_cluster': most_common_cluster,
        'most_common_cluster_count': most_common_cluster_count
    })

# Mengonversi hasil ke DataFrame
results_df = pd.DataFrame(results)

# Menampilkan hasil analisis
print(results_df)
print(df['cluster'].value_counts())

# Filter data untuk cluster 0
df_cluster_0 = df[df['cluster'] == 0]

# Lakukan clustering ulang pada data cluster 0 menjadi 60 cluster
kmeans_cluster_0 = KMeans(n_clusters=n_clusters, random_state=42)
clusters_cluster_0 = kmeans_cluster_0.fit_predict(df_cluster_0.drop(columns=[major_column, 'cluster']))

# Menambahkan informasi cluster baru ke dalam DataFrame
df_cluster_0['cluster_new'] = clusters_cluster_0

# Menampilkan hasil clustering ulang
print(df_cluster_0['cluster_new'].value_counts())
```
