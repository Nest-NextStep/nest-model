```python
major_counts = df['major'].value_counts()
print(major_counts)
```

```python
majority_class = major_counts[major_counts > 1000]
```

```python
max_instances = 400
balanced_dfs = []
for major, count in major_counts.items():
    if count > max_instances:
        sampled_df = df[df['major'] == major].sample(n=max_instances, random_state=42)
    else:
        sampled_df = df[df['major'] == major]
    balanced_dfs.append(sampled_df)

# Combine all the samples into a single DataFrame
balanced_df = pd.concat(balanced_dfs)

# Shuffle the dataset to mix the classes
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Check the new class distribution
print("Balanced class distribution:\n", balanced_df['major'].value_counts())
```
